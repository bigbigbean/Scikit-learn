```python
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn import datasets
from sklearn.linear_model import LinearRegression

#高度统一的形式
loaded_data = datasets.load_boston()
data_X = loaded_data.data    #.data就是x
data_y = loaded_data.target  #.traget就是y

model = LinearRegression()  #选择model
model.fit(data_X,data_y)    #model训练完成

print(model.predict(data_X[:4,:]))
print(data_y[:4])           #前四个数据的预测值与真实值的比较
```

前四个数据的预测值与真实值的比较，可见只是较接近，还需要很多优化。
以上是load已有数据集的方法，下面是创造一些数据的方法。

```python
import matplotlib.pyplot as plt

X,y = datasets.make_regression(n_samples = 100, n_features =1 , n_targets = 1,noise =100)
plt.scatter(X,y)
plt.show()
```

noise可以调整生成更为集中或离散的点。上面noise=100

```python
import matplotlib.pyplot as plt

X,y = datasets.make_regression(n_samples = 100, n_features =1 , n_targets = 1,noise =10)
plt.scatter(X,y)
plt.show()
```
